```python
import networkx as nx
import matplotlib.pyplot as plt

# Define the graph
G = nx.DiGraph()
G.add_edges_from([('A1', 'A2'), ('A1', 'A3'), ('A2', 'A3'), ('A3', 'A5'), ('A4', 'A5'), ('A5', 'A4'), ('A5', 'A6')])

# Visualize the graph
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_size=500)
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos, font_size=20, font_family="sans-serif")
plt.axis('off')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create an empty directed graph
G = nx.DiGraph()

# Add edges to the graph
edges = [("A1", "A2"), ("A2", "A1"), ("A2", "A3"), ("A3", "A4"), ("A4", "A5"), ("A5", "A4"), ("A5", "A3")]
G.add_edges_from(edges)

# Visualize the graph
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=12, edge_color='gray')
plt.axis('off')
plt.show()
```
