```python
import networkx as nx
import matplotlib.pyplot as plt

def construct_disconnected_graph(n):
    G = nx.Graph()
    
    # Create n isolated vertices
    G.add_nodes_from(range(n))
    
    # Create n-1/2 edges within the first (n-1)/2 vertices
    for i in range((n-1)//2):
        G.add_edge(i, i+1)
    
    return G

# Example usage
n = 6
G = construct_disconnected_graph(n)

# Visualize the graph
pos = nx.spring_layout(G)
plt.figure(figsize=(6, 6))
nx.draw_networkx(G, pos, with_labels=True, node_size=500, node_color='lightblue')
plt.title("Disconnected Graph")
plt.show()
```
