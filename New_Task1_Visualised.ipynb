```python
# Define the DFS algorithm
def DFS(graph, start):
    visited = set() 
    stack = [start]
    while stack:
        vertex = stack.pop()
        if vertex not in visited:
            visited.add(vertex)
            stack = list((graph.get(vertex, set()) - visited)) + stack
    return visited

# Define the graph
graph = {
    1: {7, 3},
    2: {2},
    3: {8, 5, 4},
    4: {1, 4},
    5: {9}
}

# Call DFS on the graph starting at node 1
visited_nodes = DFS(graph, 1)

# Print the visited nodes
print("Visited nodes:", visited_nodes)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Define the graph
graph = {
    1: {7, 3},
    2: {2},
    3: {8, 5, 4},
    4: {1, 4},
    5: {9}
}

# Create a directed graph object from the dictionary
G = nx.DiGraph(graph)

# Draw the graph using a circular layout
pos = nx.circular_layout(G)
nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos)

# Show the plot
plt.axis('off')
plt.show()
```
