```python


import networkx as nx
import matplotlib.pyplot as plt

# Step 1: Define the incoming train interval
sigma = [(1, 2), (2, 3), (3, 4), (4, 5), (3, 6), (1, 7), (4, 8), (5, 9), (4, 10), (1, 11)]

# Step 2: Create the interval graph
G = nx.Graph()
for i, interval in enumerate(sigma):
    G.add_node(i, interval=interval)
    for j in range(i):
        if interval[1] > sigma[j][0] and interval[0] < sigma[j][1]:
            G.add_edge(i, j)

# Step 3: Find the minimum coloring of the graph
color_map = nx.greedy_color(G, strategy="largest_first")

# Step 4: Visualize the graph and its coloring
pos = {i: ((interval[0] + interval[1]) / 2, -color_map[i]) for i, interval in nx.get_node_attributes(G, "interval").items()}
nx.draw(G, pos, with_labels=True, node_color=list(color_map.values()), cmap=plt.cm.rainbow)
plt.show()
```
